```python
# 
import pyspark.sql.functions as F
from pyspark.sql.types import *
```

```python
def get_total(*columns):
    total = 0
    for col_value in columns:
        total += col_value
    return total
udf_get_total = F.udf(get_total,DoubleType())
```

```python
# pyspark dataframe colume  type (): str -> int or float, double
rback_cols: list = rback.columns  
    
for i in range(1,len(rback_cols)):
    col = rback_cols[i]
    if i ==1:
        new_rbackDF = rback.withColumn(col, rback[col].cast(DoubleType()))
    else:
        new_rbackDF = new_rbackDF.withColumn(col, new_rbackDF[col].cast(DoubleType()))

        
new_rbackDF
```

```python
new_rbackDF = new_rbackDF.select([c for c in new_rbackDF.columns if c not in {'Date'}])
```

```python
new_rbackDF.withColumn('sum',udf_get_total(*[col for col in new_rbackDF.columns])).show()
```
